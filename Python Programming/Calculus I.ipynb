## Segment 2: Computing Derivatives with Differentiation

Let's bring back our ol' buddy y = x**2 + 2*x + 2:

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```

Let's identify the slope where, say x = 2

First, let's determinte what y is:

```python
f(2)
```

Cool, let's call this point P, which is located at (2, 10):

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2,10)
_ = ax.plot(x,y)
```

The delta method uses the difference between two points to calculate slope.

To illustrate this, let's define another point, Q where, say x = 5

```python
f(5)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2,10)
plt.scatter(5,37, c='orange', zorder=3)
_ = ax.plot(x,y)
```

To find the slope m between points P and Q:

m = delta(y)/delta(x) = (37-10)/(5-2) = 9

```python
m = (37-10)/(5-2)
m
```

To plot the line that passes through P and Q, we can rearrange the equation of a line y = mx + b to solve for b:

b = y - mx

```python
b = 37 - m*5
b
```

```python
b = 10 - m*2
b
```

```python
line_y = m*x + b
line_y
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2,10)
plt.scatter(5,37, c='orange', zorder=3)
plt.ylim(-5,150)
plt.plot(x,line_y,c='orange')
_ = ax.plot(x,y)
```

The closer Q becomes to P, the closer the slope m comes to being the true tangent of the point P.

Let's demonstrate this with another point Q at x = 2.1

Previously, our delta(x) between Q and P was equal to 3. Now it is much smaller:

delta(x) = x2 - x1 = 2.1 - 2 = 0.1

```python
f(2.1)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2,10)
plt.scatter(2.1,10.61, c='orange', zorder=3)
_ = ax.plot(x,y)
```

```python
m = (10.61 - 10) / (2.1 - 2)
m
```

```python
b = 10.61 - m*2.1
b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2,10)
plt.scatter(2.1,10.61, c='orange', zorder=3)
plt.ylim(-5,150)
plt.plot(x,line_y,c='orange', zorder=3)
_ = ax.plot(x,y)
```

The closer Q becomes to P (i.e., delta(x) approaches 0), the clearer it becomes that the slope m at point P = (2,10) is equal to 6.

Let's make delta(x) extremely small, 0.000001, to illustrate this:

```python
delta_x = 0.000001
delta_x
```

```python
x1=2
y1=10
```

Rearranging delta(x) = x2 - x1, we can calculate x2 for our point Q, which is now extremely close to P:

x2 = x1 + delta(x)

```python
x2 = x1 + delta_x
x2
```

y2 for our point Q can be obtained with the usual function f(x):

y2 = f(x2)

```python
y2 = f(x2)
y2
```

To find the slope, we continue to use

m = delta(y) / delta(x)

```python
m = (y2 - y1)/(x2 - x1)
m
```

Boom! using the delta method, we've shown that at point P, the slope of the curve is 6

```python
x1 = -1
delta_x = 0.000001
x2 = x1 + delta_x
x2
```

```python
def f(my_x):
    my_y = my_x ** 2 + 2 * my_x + 2
    return my_y
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y = f(x)
```

```python
y1 = f(x1)
y1
```

```python
y2 = f(x2)
y2
```

```python
m = (y2 - y1) / (x2 - x1)
m
```

```python
b = y1 - m * x1
b
```

```python
line_y = m * x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, c='lightgray')
plt.axhline(y=0, c='lightgray')
plt.xlim(-10, 10)
plt.ylim(-1, 100)
ax.scatter(x1, y1, c='green')
ax.scatter(x2, y2, c='orange', zorder=3)
plt.plot(x, line_y, c='orange', zorder=3)
_ = ax.plot(x, y, c='green')
```

As Q becomes infinitely close to P:
- x2 - x1 approaches 0
- In other words, delta(x) approaches 0
- This can be denoted as delta(x) -> 0

Using the delta method, we've derived the definition of differentiation from first principles.
The derivative of y (denoted dy) with respect to x (denoted dx) can be represented as:

dy/dx = lim delta(x)->0 ( delta(y) / delta(x) )

Expanding delta(y) out to y2 - y1:

dy/dx = lim delta(x)->0 ( (y2-y1) / delta(x) )

Finally, replacing y1 with f(x) and replacing y2 with f(x + delta(x)),
we obtain a common representation of differentiation:

dy/dx = lim delta(x)->0 ( ( f(x + delta(x)) - f(x) ) / delta(x) )

Let's observe the differentiation equation in action:

```python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(f, -1, delta))
```

## Segment 3: Automatic Differentiation

TensorFlow and PyTorch are the two most popular automatic differentiation libraries.

Let's use them to calculate dy/dx at x = 5 where:

y = x**2

dy/dx = 2*x = 2(5) = 10

### Autodiff with PyTorch

```python
import torch
```

```python
x = torch.tensor(5.0)
x
```

```python
x.requires_grad_() # contagiously track gradients through forward pass
```

```python
y=0
```

```python
y = x**2
```

```python
y.backward() # use autodiff
```

```python
x.grad
```

### Autodiff with TensorFlow

```python
import tensorflow as tf
```

```python
x = tf.Variable(5.0)
x
```

```python
with tf.GradientTape() as t:
    t.watch(x) # track forward pass
    y = x**2
```

```python
t.gradient(y, x) # use autodiff
```
